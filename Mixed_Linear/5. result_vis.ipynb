```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 한글 시각화 설정
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus']= False

path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=path).get_name()
plt.style.use('seaborn-darkgrid') # https://python-graph-gallery.com/199-matplotlib-style-sheets/
rc('font', family=font_name)
```

```python
plt.style.use('ggplot')
```

```python
df = pd.read_csv('./Data_Result(결과 데이터)/result_local.csv')
```

```python
df.set_index('date', inplace=True, drop=True)
```

```python
plt.figure(figsize=(15,15))

ax1 = plt.subplot(2,1,1)
plt.plot(df['현재수요(MW)_x'], label='실제 전력수요')
plt.plot(df['현재수요(MW)_y'].values, label='예측 전력수요')
plt.title("실제 전력수요와 예측 전력수요")
plt.ylabel("전력수요")
plt.ylim([40000,75000])
plt.xticks(np.arange(0,len(df)+1,144) , rotation=15)
plt.legend()

ax2 = plt.subplot(2,1,2, sharex=ax1)
plt.plot(df['Error(실제-예측)'], color='tab:orange')
plt.title("실제값과 예측값의 차이")
plt.axhline(0, color='gray', linestyle='--', linewidth=2)
plt.ylabel("<- 예측 수요 우세               실제 수요 우세 ->")
plt.ylim([-15000,15000])
plt.xticks(np.arange(0,len(df)+1,144), rotation=15)
plt.show()
```
